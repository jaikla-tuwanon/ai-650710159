1.1 ทำให้ข้อมูลเป็นมาตรฐาน

```python
def standardize_faculty(เพศ):
    if เพศ in ['M', 'ชาย', 'Male', 'ช']:
        return 'M'
    elif เพศ in ['F', 'หญิง', 'Female','ญ']:
        return 'F'
    else:
        return เพศ

# ใช้งาน
df['เพศ_มาตรฐาน'] = df['เพศ'].apply(standardize_faculty)
```

```python
def standardize_faculty(คณะ):
    if คณะ in ['วิทย์', 'วิทยาศาสตร์', 'Science','คณะวิทยาศาสตร์','Sci']:
        return 'วิทย์'
    elif คณะ in ['ศิลป์', 'Arts', 'ศิลปศาสตร์','Liberal Arts']:
        return 'ศิลป์'
    elif คณะ in ['วิศวะ', 'Engineering', 'วิศวกรรม','วิศวฯ','วิศวกรรมศาสตร์']:
        return 'วิศวะ'
    else:
        return คณะ

# ใช้งาน
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

1.2 เติมค่าด้วยค่าที่พบบ่อยที่สุด

```python
# เติมค่าว่างในคอลัมน์ 'เพศ_มาตรฐาน' ด้วยค่าที่พบบ่อยที่สุด
df['เพศ_มาตรฐาน'] = df['เพศ_มาตรฐาน'].fillna(df['เพศ_มาตรฐาน'].mode()[0])
```

```python
# เติมค่าว่างในคอลัมน์ 'คณะ_มาตรฐาน' ด้วยค่าที่พบบ่อยที่สุด
df['คณะ_มาตรฐาน'] = df['คณะ_มาตรฐาน'].fillna(df['คณะ_มาตรฐาน'].mode()[0])
```

-การจัดการค่าผิดปกติ

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")

# ใช้ IQR (Interquartile Range)
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# กรองค่าผิดปกติ
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
```

การรวม Features

```python
# สร้าง DataFrame สำหรับ Feature ด้านสุขภาพ
student_health = df[['ชื่อ', 'ส่วนสูง_cm', 'น้ำหนัก_kg']].copy()

# สร้าง BMI
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```
